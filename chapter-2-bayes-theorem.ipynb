```python
import pandas as pd
from fractions import Fraction

def update_posterior(tbl):
    tbl['unnorm'] = tbl.prior * tbl.likelihood
    prob = tbl['unnorm'].sum()
    tbl['posterior'] = tbl['unnorm'] / prob
    return prob

```

```python
# 2 bowls problem

tbl = pd.DataFrame(index=['Bowl 1', 'Bowl 2'])
tbl['prior'] = 1/2, 1/2
tbl['likelihood'] = 3/4, 1/2
unnorm_prob = update_posterior(tbl)

tbl

```

```python
# dice problem

tbl = pd.DataFrame(index=['6 Sided | x=1','8 Sided | x=1','12 Sided | x=1'])
tbl.index.name = 'hypothesis'
tbl['prior'] = Fraction(1, 3), Fraction(1, 3), Fraction(1, 3)
tbl['likelihood'] = Fraction(1, 6), Fraction(1, 8), Fraction(1, 12)
unnorm_prob = update_posterior(tbl)

tbl

```

```python
# normal trick coin problem

tbl = pd.DataFrame(index=['Normal | x=H','Trick Coin | x=H'])
tbl.index.name = 'hypothesis'
tbl['prior'] = Fraction(1, 2), Fraction(1, 2)
tbl['likelihood'] = Fraction(1, 2), Fraction(1, 1)
unnorm_prob = update_posterior(tbl)

tbl

```

```python
# two kids, girls boys problem

tbl = pd.DataFrame(index=['BB | G>=1','BG | G>=1','GG | G>=1','GB | G>=1'])
tbl.index.name = 'hypothesis'
tbl['prior'] = Fraction(1, 4), Fraction(1, 4), Fraction(1, 4), Fraction(1, 4)
tbl['likelihood'] = 0, 1, 1, 1
unnorm_prob = update_posterior(tbl)

tbl
```
